**DAta Preprocessing

```python
x=data.drop(["Churn"],axis=1)
y=data["Churn"]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x['TotalCharges']=scaler.fit_transform(x[['TotalCharges']])
x['tenure']=scaler.fit_transform(x[['tenure']])
x['MonthlyCharges']=scaler.fit_transform(x[['MonthlyCharges']])
x.head()
```

```python
categorical.remove('Churn')
x[categorical].head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for c in categorical:
  x[c]=encoder.fit_transform(x[c])
  mapping= dict(zip(encoder.classes_,encoder.transform(encoder.classes_)))
  print(c,":",mapping)
```

```python
x
```

```python
x.corr()
```

```python
temp=pd.concat([x,y],axis=1)
```

```python
y=encoder.fit_transform(y)
```

```python
sns.heatmap(x.corr())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
kbest=SelectKBest(f_classif,k=10)
kbest.fit(x_train,y_train)
x_train_selected=x_train[x_train.columns[kbest.get_support()]]
x_test_selected=x_test[x_test.columns[kbest.get_support()]]
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model1=DecisionTreeClassifier()
model1.fit(x_train,y_train)
ans1=model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score1= accuracy_score(y_test,ans1)
score1*100
```

```python
from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier()
model2.fit(x_train_selected,y_train)
ans2=model2.predict(x_test_selected)
score2=accuracy_score(y_test,ans2)
score2*100
```

```python
cm=confusion_matrix(y_test,ans1)
sns.heatmap(cm,annot=True,fmt='g')
```

```python
print(classification_report(y_test,ans1))
```

```python
cd=confusion_matrix(y_test,ans2)
sns.heatmap(cd,annot=True,fmt='g')
```

```python
print(classification_report(y_test,ans2))
```
